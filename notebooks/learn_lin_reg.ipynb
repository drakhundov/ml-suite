```python
from mlsuite.learn import CalHouseLearner
from mlsuite.reg import LinearRegression, OptimizationMethod, GDConfig
```

```python
config = GDConfig(lr=1e-9)
model = LinearRegression(OptimizationMethod.GRAD, gdconf=config)

learner = CalHouseLearner(model)
```

```python
learner.fit()
learner.test()
```
